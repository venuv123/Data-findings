```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
pitches = pd.read_csv('pitches.csv')
```

```python
atbats = pd.read_csv('atbats.csv')
```

```python
player_names = pd.read_csv('player_names.csv')
```

```python
games = pd.read_csv('games.csv')
```

```python
pitches.head()
```

```python
pitches = pitches.reindex(sorted(pitches.columns), axis=1)
```

```python
pitches.head()
```

```python
atbats.head()
```

```python
games.head()
```

```python
player_names.head()
```

```python
pa = pitches.merge(atbats, left_on='ab_id', right_on = 'ab_id', how = 'left')
```

```python
pa.head()
```

```python
pag = pa.merge(games, left_on = 'g_id', right_on = 'g_id', how = 'left')
```

```python
pag
```

```python
player_names['Name']=player_names['first_name'].str.cat(player_names['last_name'], sep=" ")
```

```python
player_names.drop(columns = ['first_name', 'last_name'], inplace = True)
```

```python
player_names.head()
```

```python
p = player_names.set_index('id')['Name'].to_dict()
v = pag.filter(like = 'pitcher_id')
pag[v.columns] = v.replace(p)
```

```python
p = player_names.set_index('id')['Name'].to_dict()
v = pag.filter(like = 'batter_id')
pag[v.columns] = v.replace(p)
```

```python
pag
```

```python
pag = pag.query('pitcher_id == "Zack Greinke" and pitch_type != "IN" and start_speed != " "')
```

```python
pag.to_csv('pitch_data_merged.csv', index = False)
```
